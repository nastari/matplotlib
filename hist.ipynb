```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.style.use("seaborn-white")
%matplotlib inline
```

```python
x1 = np.random.normal(-3,1,1000) # (mean, std , size )
x2 = np.random.normal(0,2,1000)
x3 = np.random.normal(5,1,1000)
```

```python
plt.figure( figsize = (15,10))
kwargs = dict( histtype = 'stepfilled', alpha = 0.3, bins = 40 )
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

```python
plt.figure( figsize = (15,10))
kwargs = dict( histtype = 'stepfilled', alpha = 0.3, bins = 40, density = True )

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
# Density nos da as probabilidades
plt.title('Probabilities each point in each histogram')
```

# Multivaloriate Histogram

```python
mean = [0,0]
cov  = [[1,1],[1,2]]
plt.figure( figsize = (15,10))
x,y = np.random.multivariate_normal(mean,cov,10000).T
plt.hist2d(x,y,bins = 30, cmap='Blues' )
cb = plt.colorbar()
cb.set_label('count in bins')
```

```python
plt.figure( figsize = (15,10))
plt.hexbin( x,y, gridsize= 35, cmap = 'Blues')
cb = plt.colorbar(label='count in bin')
```

```python
from scipy.stats import gaussian_kde
plt.figure( figsize = (15,10))
data = np.vstack([x, y])
kde = gaussian_kde(data)
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))
# Plot the result as an image
plt.imshow(Z.reshape(Xgrid.shape),
origin='lower', aspect='auto',
extent=[-3.5, 3.5, -6, 6],
cmap='Blues')
cb = plt.colorbar()
cb.set_label("density")
```
